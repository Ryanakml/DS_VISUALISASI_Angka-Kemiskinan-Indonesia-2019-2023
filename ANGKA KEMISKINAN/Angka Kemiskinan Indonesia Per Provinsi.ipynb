```python
dff = pd.DataFrame({})
```

```python
# Data Frame Baru

dff['Wilayah'] = df.iloc[:,0]

dff['2015'] = (df.iloc[:,1] + df.iloc[:,2])/2
dff['2016'] = (df.iloc[:,3] + df.iloc[:,4])/2
dff['2017'] = (df.iloc[:,5] + df.iloc[:,6])/2
dff['2018'] = (df.iloc[:,7] + df.iloc[:,8])/2
dff['2019'] = (df.iloc[:,9] + df.iloc[:,10])/2
dff['2020'] = (df.iloc[:,11] + df.iloc[:,12])/2
dff['2021'] = (df.iloc[:,13] + df.iloc[:,14])/2
dff['2022'] = (df.iloc[:,15] + df.iloc[:,16])/2
dff['2023'] = df.iloc[:,17]
```

```python
dff.dropna(inplace=True)
dff.columns
```

# Visualisasi pertumbuhan per tahun

```python
indo = dff.tail(1)
indo_th = indo.drop(columns='Wilayah')
x = indo_th.columns
y = indo_th.values.flatten()
y
```

```python
indo_th.columns
x = pd.Index(x).values
x
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Misalkan 'indo_th' adalah DataFrame yang berisi data yang akan diplot

# Mengambil nama kolom sebagai sumbu x dan nilai dari kolom 'Wilayah' sebagai sumbu y
x = indo_th.columns.tolist()  # Mengambil nama kolom sebagai sumbu x
y = indo_th.values.flatten()   # Mengambil nilai dari kolom 'Wilayah' sebagai sumbu y

# Membuat plot menggunakan seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=x, y=y, palette='viridis')
plt.title('Angka Kemiskinan di Indonesia per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Angka Kemiskinan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
